```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read train file

```python
df = pd.read_csv('//workspaces//NYC_taxi_trip//data//raw//train//train.csv')
df.head()
```

```python
df.info()
```

```python
# convert datetime object to 'Datetime' format
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime']) 
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime']) 
```

```python
# extract the day, month, and year components (for pickup)
df['pickup_day'] = df['pickup_datetime'].dt.day
df['pickup_month'] = df['pickup_datetime'].dt.month
df['pickup_year'] = df['pickup_datetime'].dt.year
df['pickup_Hour'] = df['pickup_datetime'].dt.hour
df['pickup_Minute'] = df['pickup_datetime'].dt.minute
df['pickup_Second'] = df['pickup_datetime'].dt.second

# extract the day, month, and year components (for dropup)
df['dropoff_day'] = df['dropoff_datetime'].dt.day
df['dropoff_month'] = df['dropoff_datetime'].dt.month
df['dropoff_year'] = df['dropoff_datetime'].dt.year
df['dropoff_Hour'] = df['dropoff_datetime'].dt.hour
df['dropoff_Minute'] = df['dropoff_datetime'].dt.minute
df['dropoff_Second'] = df['dropoff_datetime'].dt.second
```

```python
df.head()
```

```python
# Drop  'pickup_datetime' and 'dropoff_datetime' column form DataFrame as it converted to individual feature 
df.drop(['id','pickup_datetime', 'dropoff_datetime'],
        axis=1,
        inplace=True)
```

```python
# replace on  col   'store_and_fwd_flag'
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].replace({'N':0, 'Y':1})
```
